### Import & Load

```python
import json
import seaborn as sns
import matplotlib.pyplot as plt
plt.rc('font', family='NanumBarunGothic')

with open('importances_dict.json', 'r') as json_file:
    importances_dict = json.load(json_file)
```

### Setting

```python
columns_list = ['법정동', '층', '건축년도', '용적률', '건폐율', '노령화지수', '전세가율', '소비자물가지수', '회사채', '지하철역 개수']

def show_importance(name_of_location):
  sns.barplot(x = importances_dict[name_of_location], y = columns_list).set_title(name_of_location)
  plt.show()
```

### 실행

```python
show_importance('강남구')
```
