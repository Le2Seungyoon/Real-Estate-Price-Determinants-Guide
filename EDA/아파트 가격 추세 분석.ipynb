```python
df = df[["Unnamed: 0", "일반평균가"]]
```

```python
df.rename(columns = {"Unnamed: 0" : "Time"}, inplace = True)
df.rename(columns = {"일반평균가" : "Average Sale Price"}, inplace = True)
```

```python
df["Price Shift"] = df["Average Sale Price"].shift(12)
```

```python
df["Change Rate"] = ((df["Price Shift"] - df["Average Sale Price"]) / df["Average Sale Price"])*100 
```

시각화

```python
df.plot(x='Time', y='Average Sale Price') #가격 추세 그래프
```

```python
df.plot(x='Time', y='Change Rate') #변화율 그래프
```

```python
import matplotlib.pyplot as plt #시각화 모듈

#그래프 겹쳐 그리기
fig, ax1 = plt.subplots()
ax1.plot(df["Time"], df["Average Sale Price"], color="red")

ax2 = ax1.twinx() #그래프 겹치기
ax2.plot(df["Time"], df["Change Rate"], color="blue")

plt.show()
```
