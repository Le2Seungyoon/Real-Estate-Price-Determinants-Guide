```python
import pandas as pd
```

```python
df=pd.read_csv("./2021실거래가.csv",encoding='CP949')
```

```python
df.head()
```

```python
df.dropna(axis=0,subset=['본번','부번'],inplace=True)
```

```python
df.astype({'본번':int,'부번':int})
```

```python
df.insert(0,'계약일자',df['계약년월']*100+df['계약일'])
##df['계약일']=df['계약년월']*100+df['계약일']
df.drop(labels='계약년월',axis=1,inplace=True)
df.drop(labels='계약일',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.insert(1,'구',df['시군구'])
df.insert(2,'동',df['시군구'])
df.drop(labels='시군구',axis=1,inplace=True)
df.drop(labels='중개사소재지',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
##여기 범위에 숫자만 info 보고 숫자 넣어 주세요
for i in range(0,43413):
    str1=df.iloc[i,1]
    str2=str1.split()
    df.iloc[i,1]=str2[1]
    df.iloc[i,2]=str2[2]
    str3=df.iloc[i,8]
    str4=str3.split(',')
    df.iloc[i,8]="".join(str4)
```

```python
df.astype({"거래금액(만원)":int})
```
