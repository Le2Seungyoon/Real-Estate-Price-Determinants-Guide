```python
import pandas as pd
```

```python
df_train = pd.read_csv("./서울시 지하철역 연계 지하도 공간정보.csv", encoding='cp949')
```

```python
df_train = df_train[["시군구코드", "시군구명", "읍면동코드", "읍면동명", "지하철역명"]]
```

```python
df_train
```

```python
df_train=df_train.drop_duplicates()
df_train=df_train.drop_duplicates(['읍면동코드', '지하철역명'])
```

```python
df_train
```

```python
df_counts=df_train["읍면동명"].value_counts().to_frame()
```

```python
df_counts=df_counts.reset_index().rename(columns={"index":"읍면동명","읍면동명":"개수"})
```

```python
df_counts
```

```python
df=pd.merge(df_train,df_counts)
```

```python
df
```

```python
df.to_csv("동별지하철역.csv",encoding='cp949',index=False)
```

```python
df_building=pd.read_csv("./서울시 건축물대장 정보.csv", encoding='cp949',low_memory=False)
```

```python
df_building=df_building[["관리_건축물대장_PK","시군구_코드","법정동_코드"]]
df_building
```

```python
df_type=pd.read_csv("./서울시 건축물대장 주용도.csv",encoding='utf-8')
```

```python
df_type
```

```python
df_type['주용도 구분'].unique()
```

```python
df_type1=df_type[["건축물대장 관리번호","건폐율","용적율"]]
```

```python
df_type1
```

```python
df_merge=pd.merge(df_type1,df_building,right_on="관리_건축물대장_PK",left_on="건축물대장 관리번호")
```

```python
df_merge
```

```python
idx=df_merge[(df_merge['용적율']==0.0)|(df_merge['건폐율']==0.0)].index
df_m=df_merge.drop(idx)
```

```python
df_m.drop(['관리_건축물대장_PK'],axis=1,inplace=True)
```

```python
df_m
```

```python
df_m.to_csv("동별 건폐율 용적율.csv",encoding='cp949',index=False)
```
